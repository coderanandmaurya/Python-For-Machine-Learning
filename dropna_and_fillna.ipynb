```python
import pandas as pd

 #if we want to write 0 in those columns which have nan
df = pd.read_csv('https://raw.githubusercontent.com/coderanandmaurya/PANDAS/main/dropnaexample.csv')
df
```

```python
df1= df.dropna()
df1
```

## Replace all NaN elements with 0s

```python
dfz1=df.copy()
```

```python
dfz1.fillna(0, inplace= True)
dfz1
```

## Replacing NaN values with a Static value.

```python
df_fill = df.fillna('CSE')   #you can fill any number
df_fill 
```

```python
df
```

# only one col

```python
dfc=df.copy()
```

```python
dfc["Python"].fillna("CSE",inplace=True)
dfc
```

##  Using method Parameter

```python
dfmf=df.copy()
```

```python
# forward fill
dfmf["CSE"].fillna( method ='ffill', inplace = True)
dfmf
```

```python
dfmb=df.copy()
```

```python
# backfill
dfmb["CSE"].fillna( method ='bfill', inplace = True)
dfmb
```
